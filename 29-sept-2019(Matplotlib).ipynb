### It is the library function for Ploting graphs based on your Data 
We have different types of plots.they are
- plot
- pie
- bar

```python
import matplotlib.pyplot as mp

mp.plot([9,2,4],[1,5,7])
mp.show()
```

```python
k=list(range(1,10))

p=[i**2 for i in range(len(k))]

mp.plot(k,p,'r*')
```

```python
import numpy as np

x=list(range(0,30))
y=np.sin(x)
mp.plot(x,y,'g--')
```

```python
x=np.arange(0,4*np.pi,0.1)
y=np.sin(x)
mp.plot(x,y,'g*')
mp.title("Sine Graph")
mp.xlabel("X Axis")
mp.ylabel("Y Axis")
mp.grid()
```

```python
# How to merge 2 bar graphs
import matplotlib.pyplot as mp
p=list(range(6))
y=[74,18,52,9,63,35]
z=[63,25,32,0,48,26]
mp.bar(p,y)
mp.bar(p,z)
mp.legend("Plot1","Plot2")
```
